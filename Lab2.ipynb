```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv('Dataset.csv')

print(dataset.info())
```

```python
def handle_missing_data(dataset):
    """Handles missing data by filling with the mode for categorical columns."""
    if 'Profession' in dataset.columns:
        dataset['Profession'].fillna(dataset['Profession'].mode()[0], inplace=True)
    return df

dataset = handle_missing_data(dataset)
```

```python
def plot_numeric_histograms(df, numeric_columns):
    """Plots histograms for numeric columns."""
    df[numeric_columns].hist(bins=15, figsize=(15, 10), color='skyblue', edgecolor='black')
    plt.suptitle('Distribution of Numeric Variables', fontsize=16)
    plt.tight_layout()
    plt.show()

numeric_columns = ['Age', 'Annual Income', 'Spending Score', 'Work Experience', 'Family Size']
plot_numeric_histograms(dataset, numeric_columns)
```

```python
def plot_nominal_countplots(df, nominal_columns):
    """Plots count plots for nominal columns."""
    fig, axes = plt.subplots(1, len(nominal_columns), figsize=(15, 5))
    for i, col in enumerate(nominal_columns):
        sns.countplot(data=df, x=col, ax=axes[i], palette="viridis")
        axes[i].set_title(f'Distribution of {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Count')
        axes[i].tick_params(axis='x', rotation=45)
    plt.tight_layout()
    plt.show()

nominal_columns = ['Gender', 'Profession']
plot_nominal_countplots(dataset, nominal_columns)
```

```python
def plot_scatter(df, x, y):
    """Plots a scatterplot for two numeric variables."""
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=x, y=y, hue='Gender', palette='Set2', alpha=0.7)
    plt.title(f'{x} vs {y}')
    plt.show()

plot_scatter(dataset, 'Annual Income', 'Spending Score')
```

```python
def plot_correlation_heatmap(df, numeric_columns):
    """Plots a heatmap of correlations for numeric columns."""
    correlation_matrix = df[numeric_columns].corr()
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
    plt.title('Correlation Heatmap')
    plt.show()

plot_correlation_heatmap(dataset, numeric_columns)
```

```python
def plot_boxplot(df, x, y):
    """Plots a boxplot for a numeric variable grouped by a categorical variable."""
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x=x, y=y, palette="Set3")
    plt.title(f'Boxplot of {y} by {x}')
    plt.show()

plot_boxplot(dataset, 'Gender', 'Annual Income')
```

```python
def plot_violinplot(df, x, y):
    """Plots a violin plot for a numeric variable grouped by a categorical variable."""
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=df, x=x, y=y, palette="muted", inner="quartile")
    plt.title(f'Violin Plot of {y} by {x}')
    plt.show()

plot_violinplot(dataset, 'Profession', 'Spending Score')
```
